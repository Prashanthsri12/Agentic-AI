### Models Integration with OpenAI, GORQ and Google Gemini

```python
import os
from dotenv import load_dotenv

load_dotenv()

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("gpt-4.1")
response = model.invoke("Hello, How are you?")
print(response.content)
```

### Google Gemini Model Integration

```python
model = init_chat_model("google_genai:gemini-2.5-flash-lite")
response = model.invoke("why do parrots talk?")
print(response.content)
```

### ChatOpenAI


```python
from langchain_openai import ChatOpenAI

model = ChatOpenAI(model="gpt-4.1")
response = model.invoke("Hello, How are you?")
print(response.content)
```

### ChatGoogleGenerativeAI

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
response = model.invoke("What is called as the constitution of India?")
print(response.content)
```

### GORQ Model Integration

### Streaming and Batch

#### Streaming

```python
model = ChatOpenAI(model="gpt-4.1")
model.invoke("Write a 200 word essay on the constitution of India.")
```

```python
for chunk in model.stream("Write a 200 word essay on the constitution of India."):
    print(chunk.text, end="", flush=True)
```

#### Batch

```python
response = model.batch([
    "Who created the constitution of india and when was it written?",
    "When did indias first prime minister take office?",
    "Who was Tamilnadu's first chief minister?"
    ],
    config={"max_concurrency": 3})

for response in response:
    print(response)
```
