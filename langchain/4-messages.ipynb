### Message Prompts

- System Messages
- Human Messages
- AI Messages
- Tool Messages

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage, ToolMessage
```

```python
messages = [
    SystemMessage(content="You are a poetry expert."),
    HumanMessage(content="Write a thirukural about Artificial Intelligence.")
]

response = model.invoke(messages)
print(response.content)
```

```python
system_msg = SystemMessage(content="You are a helpful coding assistant.")

messages = [
    system_msg,
    HumanMessage(content="Write a python function that adds two numbers.")
]

response = model.invoke(messages)
print(response.content)
```

#### Mini Chat Space

```python
model  = init_chat_model("gpt-4.1")
messages = [
    SystemMessage(content="you are a helpful assistant, but you can only speak caveman language."),
]
while True:
    user_input = input("User: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Exiting the chat. Goodbye!")
        break
    messages.append(HumanMessage(content=user_input))
    response = model.invoke(messages)
    print(f"AI: {response.content}")
    messages.append(AIMessage(content=response.content))
```

### Tool Message

```python
ai_message = AIMessage(content="", tool_calls=[{"name":"get_weather", "args":{"location":"chennai"}, "id":"call_123"}])

weather_result = "it's 30 degrees"
tool_message = ToolMessage(content=weather_result, tool_call_id="call_123")

messages = [
    SystemMessage(content="you are a helpful assistant, but you can only speak caveman language"),
    HumanMessage(content="what's the weather in chennai?"),
    ai_message,
    tool_message
]   

model = init_chat_model("groq:qwen/qwen3-32b")
response = model.invoke(messages)
response.content
```
